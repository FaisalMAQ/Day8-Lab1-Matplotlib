```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([2,4,6,8,10])

plt.plot(pd.DataFrame([40,50,60,43,66]));
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,5), labels=['Jan','Feb','Mar', 'Apr'], explode = [.1,.1,.1,.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(300)
y = np.random.randn(300)
plt.scatter(x, y)
plt.scatter(x,y, marker='o', alpha = 0.4,s=50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title('Months', fontsize = 16, loc='left', color='r');
plt.scatter(x, y)
plt.scatter(x,y,alpha = 0.4,s=50)

plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16);

```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.plot([4,5,8,3,6])
plt.legend(['2008','2009', '2010'], loc = 2);
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])

#first
plt.subplot(2,2,1) #(number of rows, number of columns, number of plot)
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months')

#second
plt.subplot(2,2,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');

#third
plt.subplot(2,2,3) #(number of rows, number of columns, number of plot)
plt.pie(np.arange(1,5),labels=['May','Jun','Jul', 'Aug']);
plt.title('Months')
plt.plot('g')
plt.plot(c ='lime')
plt.plot(color ='blue')
```
